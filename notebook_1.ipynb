```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

print("✅ Libraries imported")
```

```python
# Load datasets
sentiment_df = pd.read_csv("fear_greed_index.csv")
trader_df = pd.read_csv("historical_data.csv")

print("Sentiment data shape:", sentiment_df.shape)
print("Trader data shape:", trader_df.shape)

sentiment_df.head(), trader_df.head()
```

```python
# Step 4: Basic sanity check (very important for beginners)
print("Sentiment columns:")
print(sentiment_df.columns)

print("\nTrader data columns:")
print(trader_df.columns)
```

```python
# PHASE 2 - STEP 1: Clean sentiment data

# Convert date column to datetime
sentiment_df["date"] = pd.to_datetime(sentiment_df["date"])

# Normalize sentiment labels (lowercase)
sentiment_df["classification"] = sentiment_df["classification"].str.lower()

# Keep only required columns
sentiment_df = sentiment_df[["date", "classification"]]

# Preview result
sentiment_df.head()
```

```python
# PHASE 2 - STEP 2: Clean trader data

# Create a copy to stay safe
trader_df = trader_df.copy()

# Convert epoch timestamp (milliseconds) to datetime
trader_df["trade_time"] = pd.to_datetime(trader_df["Timestamp"], unit="ms")

# Create trade date (to align with sentiment date)
trader_df["date"] = trader_df["trade_time"].dt.date
trader_df["date"] = pd.to_datetime(trader_df["date"])

# Rename important columns for easier use
trader_df = trader_df.rename(columns={
    "Account": "account",
    "Closed PnL": "closed_pnl",
    "Size USD": "size_usd"
})

# Remove rows without PnL (invalid or incomplete trades)
trader_df = trader_df.dropna(subset=["closed_pnl"])

# Preview cleaned data
trader_df[["account", "closed_pnl", "size_usd", "date"]].head()
```

```python
# PHASE 2 - STEP 3: Merge trader data with sentiment data

merged_df = trader_df.merge(
    sentiment_df,
    on="date",
    how="left"
)

print("Merged dataframe shape:", merged_df.shape)

# Preview merged result
merged_df[["account", "closed_pnl", "size_usd", "date", "classification"]].head()
```

```python
# PHASE 2 - STEP 4: Simplify sentiment labels

# Keep only relevant sentiment classes
valid_classes = ["fear", "extreme fear", "greed", "extreme greed"]
merged_df = merged_df[merged_df["classification"].isin(valid_classes)]

# Create simplified sentiment column
merged_df["sentiment"] = merged_df["classification"].apply(
    lambda x: "Fear" if "fear" in x else "Greed"
)

# Check distribution
merged_df["sentiment"].value_counts()
```

```python
# PHASE 3 - STEP 1: Basic trader performance metrics

# Add win/loss indicator
merged_df["is_win"] = merged_df["closed_pnl"] > 0

# Group by trader and sentiment
basic_metrics = (
    merged_df
    .groupby(["account", "sentiment"])
    .agg(
        total_pnl=("closed_pnl", "sum"),
        trade_count=("closed_pnl", "count"),
        win_rate=("is_win", "mean")
    )
    .reset_index()
)

basic_metrics.head()
```

```python
# PHASE 3 - STEP 2: ROI calculation

# Aggregate capital used (sum of trade sizes)
capital_used = (
    merged_df
    .groupby(["account", "sentiment"])
    .agg(total_capital=("size_usd", "sum"))
    .reset_index()
)

# Merge with basic metrics
metrics_with_roi = basic_metrics.merge(
    capital_used,
    on=["account", "sentiment"],
    how="left"
)

# Calculate ROI
metrics_with_roi["roi"] = (
    metrics_with_roi["total_pnl"] / metrics_with_roi["total_capital"]
)

metrics_with_roi.head()
```

```python
# PHASE 3 - STEP 3: Sharpe Ratio calculation

# Compute mean and std of PnL per trade
sharpe_df = (
    merged_df
    .groupby(["account", "sentiment"])
    .agg(
        mean_pnl=("closed_pnl", "mean"),
        std_pnl=("closed_pnl", "std")
    )
    .reset_index()
)

# Avoid division by zero
sharpe_df["sharpe_ratio"] = sharpe_df["mean_pnl"] / sharpe_df["std_pnl"]

# Merge with existing metrics
metrics_with_sharpe = metrics_with_roi.merge(
    sharpe_df[["account", "sentiment", "sharpe_ratio"]],
    on=["account", "sentiment"],
    how="left"
)

metrics_with_sharpe.head()
```

```python
# PHASE 3 - STEP 4: Maximum Drawdown calculation

def max_drawdown(pnl_series):
    """
    Computes maximum drawdown from a sequence of PnL values
    """
    cumulative = pnl_series.cumsum()
    running_max = cumulative.cummax()
    drawdown = cumulative - running_max
    return drawdown.min()

# Compute max drawdown per trader & sentiment
drawdown_df = (
    merged_df
    .sort_values("trade_time")
    .groupby(["account", "sentiment"])
    .agg(max_drawdown=("closed_pnl", max_drawdown))
    .reset_index()
)

# Merge with existing metrics
final_metrics = metrics_with_sharpe.merge(
    drawdown_df,
    on=["account", "sentiment"],
    how="left"
)

final_metrics.head()
```

```python
# PHASE 4 - STEP 1: Trader ranking by ROI

# Rank traders within each sentiment by ROI (higher is better)
final_metrics["roi_rank"] = (
    final_metrics
    .groupby("sentiment")["roi"]
    .rank(ascending=False, method="dense")
)

# Sort for easy viewing
ranked_traders = final_metrics.sort_values(
    ["sentiment", "roi_rank"]
)

# View top traders in each sentiment
ranked_traders.head(10)
```

```python
# PHASE 4 - STEP 2: Identify contrarian traders

# Pivot data so Fear and Greed metrics are side-by-side
pivot_metrics = final_metrics.pivot(
    index="account",
    columns="sentiment",
    values=["roi", "total_pnl", "sharpe_ratio"]
)

# Flatten column names
pivot_metrics.columns = [
    f"{metric}_{sentiment}"
    for metric, sentiment in pivot_metrics.columns
]

pivot_metrics = pivot_metrics.reset_index()

# Define contrarian conditions
# Good in Fear, bad in Greed OR good in Greed, bad in Fear
contrarian_traders = pivot_metrics[
    (
        (pivot_metrics["roi_Fear"] > 0) & (pivot_metrics["roi_Greed"] < 0)
    ) |
    (
        (pivot_metrics["roi_Greed"] > 0) & (pivot_metrics["roi_Fear"] < 0)
    )
]

contrarian_traders.head()
```

```python
# PHASE 4 - STEP 3: Save final metrics to CSV

output_path = "ds_ayush/csv_files/trader_metrics.csv"

final_metrics.to_csv(output_path, index=False)

print(f"✅ Final metrics saved to {output_path}")
```

```python
# PHASE 5 - STEP 1: PnL distribution by sentiment

plt.figure(figsize=(8, 5))

sns.boxplot(
    data=merged_df,
    x="sentiment",
    y="closed_pnl"
)

plt.title("PnL Distribution by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Closed PnL")

# Save plot
output_path = "ds_ayush/outputs/pnl_by_sentiment.png"
plt.savefig(output_path, bbox_inches="tight")
plt.show()

print(f"✅ Plot saved to {output_path}")
```

```python
# PHASE 5 - STEP 2: Average ROI by sentiment

# Calculate average ROI per sentiment
avg_roi = (
    final_metrics
    .groupby("sentiment")["roi"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(6, 4))

sns.barplot(
    data=avg_roi,
    x="sentiment",
    y="roi"
)

plt.title("Average ROI by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Average ROI")

# Save plot
output_path = "ds_ayush/outputs/avg_roi_by_sentiment.png"
plt.savefig(output_path, bbox_inches="tight")
plt.show()

print(f"✅ Plot saved to {output_path}")
```

```python
# PHASE 5 - STEP 3: Top traders by ROI (Fear vs Greed)

# Select top 5 traders per sentiment by ROI
top_traders = (
    final_metrics
    .sort_values("roi", ascending=False)
    .groupby("sentiment")
    .head(5)
)

plt.figure(figsize=(10, 5))

sns.barplot(
    data=top_traders,
    x="roi",
    y="account",
    hue="sentiment"
)

plt.title("Top 5 Traders by ROI under Fear vs Greed")
plt.xlabel("ROI")
plt.ylabel("Trader Account")

# Save plot
output_path = "ds_ayush/outputs/top_traders_by_roi.png"
plt.savefig(output_path, bbox_inches="tight")
plt.show()

print(f"✅ Plot saved to {output_path}")
```
